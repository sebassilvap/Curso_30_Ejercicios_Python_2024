Binario de 19 - SMALL.jpg<br>

### ***L√≥gica de los n√∫meros binarios***
- Como se puede observar de los ejemplos mostrados.
  - 19 en n√∫mero binario = 10011
  - 43 en n√∫mero binario = 101011
- La formaci√≥n del n√∫mero binario es dividiendo el n√∫mero entero para 2, de manera continua hasta que el cociente final sea 1
- Este cociente final es el punto de partida del equivalente binario
- Y a partir de este cociente se comienzan a a√±adir los residuos de cada operaci√≥n de divisi√≥n hasta llegar al cociente final.
- Obs√©rvese la direcci√≥n de la flecha en los ejemplos de arriba, con lo cual se indica el orden en el cual establecemos la formaci√≥n del n√∫mero binario.
- **RECORDAR :** Para obtener el residuo directamente de una operaci√≥n de divisi√≥n podemos utilizar el operador de m√≥dulo `%` en Python.

<br>

### ***Recordatorio sobre el operador m√≥dulo***
- B√°sicamente el m√≥dulo `%` es un operador que nos devuelve el residuo de una divisi√≥n.
- La divisi√≥n en Python es con `/`
- De manera adicional tambi√©n tenemos la divisi√≥n + floor que se obtiene con un doble operador `//`
- A continuaci√≥n se presenta un c√≥digo preparado para ejemplificar el funcionamiento b√°sico del `%`, `/` y `//`

```
# =================================================
# DATO IMPORTANTE 1
# RECORDAR EL FUNCIONAMIENTO DEL OPERADOR M√ìDULO %
# =================================================

# % => devuelve el residuo de una divisi√≥n
# EJ:
"""
9   |____5
4        1
"""

# Las partes de la divisi√≥n ser√≠an:
# Dividendo = 9
# Divisor = 5
# Cociente = 1
# Residuo (M√≥dulo) = 4

# Esto lo podemos obtener con %
print( 9 / 5 )  # divisi√≥n normal => resultado float
print( 9 // 5 ) # divisi√≥n + floor
print( 9 % 5 )  # m√≥dulo √≥ residuo
```

- La ejecuci√≥n de este c√≥digo nos dar√≠a el siguiente resultado:

```
1.8
1
4
```

- **NOTA IMPORTANTE :** Cuando el DIVIDENDO es mayor al DIVISOR, el operador m√≥dulo `%` nos devuelve el DIVIDENDO

```
# ==================================
# DATO IMPORTANTE 2
# GENERALIDADES SOBRE EL OPERADOR %
# ==================================

# ¬øqu√© pasar√≠a si intentamos 3 % 5?

print( 3 / 5 ) # 0.6
print( 3 % 5 ) # 3

# => cuando el dividendo es menor que el divisor
# => el m√≥dulo es igual al dividendo
```

- El anterior c√≥digo nos dar√≠a el siguiente resultado:

```
0.6
3
```

```
# ===========================================
# REPASO SOBRE VARIABLES GLOBALES Y LOCALES
# ===========================================

variable_1 = 10 # variable global
variable_2 = 50 # variable global

print('VARIABLES INICIALES:')
print( 'variable_1 =', variable_1 )
print( 'variable_2 =', variable_2 )

def funcion_cambiar_1():
  variable_1 = 100 # variable local diferente a variable_1 definida afuera
# end def

def funcion_cambiar_2():
  global variable_2
  variable_2 = 500 # variable global "variable_2" a la cual podemos acceder gracias a keyword "global"
# end def

print('\nDESPU√âS DE DEFINIR FUNCIONES SIN EJECUTAR:')
print( 'variable_1 =', variable_1 )
print( 'variable_2 =', variable_2 )

funcion_cambiar_1()
funcion_cambiar_2()

print('\nDESPU√âS DE DEFINIR EJECUTAR FUNCIONES:')
print( 'variable_1 =', variable_1 )
print( 'variable_2 =', variable_2 )
```

- B√°sicamente lo que tenemos que recordar de este repaso son los siguientes conceptos importantes:
  - `VARIABLE GLOBAL` : Aquella que se declara en el espectro global de nuestro c√≥digo con indentaci√≥n 0 fuera de un bloque de funci√≥n.
  - `VARIABLE LOCAL` : Aquella que se declara dentro del bloque de la funci√≥n y existe √∫nica y exclusivamente dentro de la funci√≥n. Puede tener el mismo nombre que una variable ah√≠ fuera, pero no tiene nada que ver con esta.
  - `KEYWORD "global"` : Permite el acceso a una variable global ah√≠ fuera dentro del bloque de una funci√≥n, e incluso podemos modificar esta variable dentro de la funci√≥n.

### ***C√°lculo de la Media Aritm√©tica / Promedio***

<br>

- Es el simple promedio aritm√©tico
- Por ejemplo, si se tiene una lista:

```
lista = [1, 5, 3, 2]
```

- el promedio es igual a:

```
promedio = (1 + 5 + 3 + 2) / 4
promedio = 2.75
```

<br>
<br>

### ***C√°lculo de la Mediana Estad√≠stica***

<br>

- La mediana no es m√°s que el dato que se encuentra en la posici√≥n central de una lista de datos num√©ricos ordenados de menor a mayor.
- Esto va a ser distinto, si tenemos un n√∫mero PAR de elementos y un n√∫mero IMPAR.
- Ejemplo:

```
conjunto_datos_1 = [1, 5, 3, 2]
conjunto_datos_2 = [1, 5, 3, 2, 8]
```

- Ordenando los datos de menor a mayor:

```
datos_ordenados_1 = [1, 2, 3, 5]    # n√∫mero PAR de datos
datos_ordenados_2 = [1, 2, 3, 5, 8] # n√∫mero IMPAR de datos
```

- C√≥mo se puede ver para un n√∫mero IMPAR de datos, se tiene una posici√≥n central, en este caso en el index = 2, por tanto:

```
datos_ordenados_2 = [1, 2, 3, 5, 8] # n√∫mero IMPAR de datos
#                    0  1  2  3  4

posicion_central = 2
mediana = 3
```

- Para un n√∫mero de datos IMPAR, se tienen 2 posiciones centrales y 2 valores de la mediana, y la mediana final ser√≠a el promedio de ambas

```
datos_ordenados_1 = [1, 2, 3, 5]    # n√∫mero PAR de datos
#                    0  1  2  3

posicion_central_1 = 1
posicion_central_2 = 2

mediana_1 = 2
mediana_2 = 3

mediana = (2+3)/2 = 2.5
```

<br>
<br>

### ***C√°lculo de la Moda Estad√≠stica***

<br>

- En palabras sencillas, la moda estad√≠stica es el dato que m√°s se repite en un conjunto de datos.
- Por ejemplo, para el siguiente conjunto de datos:

```
conjunto_datos = [1, 5, 2, 9, 2, 5, 7, 8, 6, 5, 3]
#                    x           x           x

moda = 5
```

- El n√∫mero 5 se repite 3 veces y el n√∫mero 2 se repite 2 veces.
- El n√∫mero 5 es el que m√°s se repite, y este ser√≠a el valor de la moda.

- Tambi√©n puede darse, que tenemos un 2 o m√°s datos que se repiten igual n√∫mero de veces, por ejemplo:

```
conjunto_datos = [1, 5, 2, 9, 2, 5, 7, 8, 6, 5, 3, 2]
#                    x           x           x
#                       o     o                    o
```

- El 5 y el 2 se repiten 3 veces
- La moda vendr√≠a a ser el promedio de ambos

```
moda = (5 + 2)/2
moda = 3.5
```

### ***TEOR√çA DE REPASO: Funciones *args***
- Recordar que una funci√≥n `*args` es aquella que:
  - Recibe un n√∫mero indefinido de par√°metros.
  - Y los devuelve como tupla
  - El nombre "args" no es importante, pero si lo es el `*`
- Un peque√±o resumen de este funcionamiento lo podemos ver en el siguiente c√≥digo:

```
# ================================================
# UN POCO DE TEOR√çA - PARTE 1
# Recordando la funci√≥n con *args
# ================================================

# ==> un ejemplo b√°sico de funci√≥n *args
def sumar(*args):
  resultado = 0

  for numero in args:
    resultado += numero
  # end for

  return resultado
# end def

# TEST
print( sumar(1,2,3,4,5) ) # 15
print( sumar(10,20) ) # 30


# ==> RECORDAR: ¬øqu√© devuelve *args?
# (el nombre args no es importante)

def que_devuelve( *argumentos ):
  print( argumentos , type(argumentos) )
# end def

# TEST
que_devuelve('hola', 10, True) # ('hola', 10, True) <class 'tuple'>
```

- En el anterior c√≥digo ya se pone en forma de comentario el resultado de cada test, sin embargo, una vez m√°s se presenta aqu√≠ cu√°l ser√≠a el resultado al ejecutar el anterior c√≥digo:

```
15
30
('hola', 10, True) <class 'tuple'>
```

- De esta manera, espero de coraz√≥n haya quedado clara la funcionalidad de las funciones args:
  - Reciben un n√∫mero indeterminado de par√°metros, puede ser 1 puede ser N.
  - El nombre *args no es relevante pero si el uso del aster√≠sco
  - B√°sicamente todos estos argumentos se almacenan a manera de tupla y podemos recorrerlos

- Habiendo recordado como operan las funciones `*args`.

def get_nombres_variables( *args ):
  tupla_nombres_variables = argname( nameof(args) )
  return tupla_nombres_variables
# end def
